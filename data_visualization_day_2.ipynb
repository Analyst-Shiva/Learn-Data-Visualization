##### 1. Write a Python program to display a bar chart of the fare of class.

```python
import matplotlib.pyplot as plt

# Grouping by 'class' and calculating the mean fare for each class
class_fare = data.groupby('class')['fare'].mean()

# Plotting the bar chart
class_fare.plot(kind='bar', color='skyblue')
plt.title('Average Fare by Class')
plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.show()
```

######5. Write a Python program to display a count plot of all categorical columns.

```python
import seaborn as sns

# Selecting only categorical columns for the count plot
categorical_columns = data.select_dtypes(include='object').columns

# Plotting count plots for each categorical column
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=column, data=data, palette='viridis')
    plt.title(f'Count Plot of {column}')
    plt.show()
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

```python
# Creating a factor plot with pclass vs embarked using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='pclass', hue='embarked', multiple='stack', bins=range(1, 5), palette='muted', element='step')
plt.title('Factor Plot of pclass vs embarked')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

###### 7. Write a Python programming to display a factor plot of ( pclass vs sex). hist (kind = 'count')

```python
# Creating a factor plot with pclass vs sex using a count histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='pclass', hue='sex', multiple='stack', bins=range(1, 5), palette='pastel', element='step')
plt.title('Factor Plot of pclass vs sex')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

###### 8. Write a Python programming to display a factor plot of ( embarked vs sex). hist (kind = 'count')

```python
# Creating a factor plot with embarked vs sex using a count histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='embarked', hue='sex', multiple='stack', palette='pastel', element='step', stat='count')
plt.title('Factor Plot of embarked vs sex')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show() 
```

###### 9. Write a Python program to display distribution plot for all numberical columns.

```python
# Selecting only numerical columns for distribution plots
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Plotting distribution plots for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=True, color='skyblue')
    plt.title(f'Distribution Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

###### 10. Write a Python program to display boxplot of age, pclass and hue = gender.

```python
# Creating a boxplot of age, pclass, and hue=sex
plt.figure(figsize=(12, 8))
sns.boxplot(x='pclass', y='age', hue='sex', data=data, palette='pastel')
plt.title('Boxplot of Age by Pclass and Gender')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()
```

##### Bonus Question

##### 1. What are the plots we can create on numerical columns.

Histograms:

Display the distribution of a single variable.
Use the plt.hist() function or Seaborn's sns.histplot().

Box Plots (Box-and-Whisker Plots):

Show the summary statistics of a variable, including median, quartiles, and potential outliers.
Use the plt.boxplot() function or Seaborn's sns.boxplot().

Violin Plots:

Combine aspects of box plots and kernel density estimation (KDE) plots to show the distribution of a variable across different categories.
Use Seaborn's sns.violinplot().

Kernel Density Estimation (KDE) Plots:

Estimate the probability density function of a variable.
Use Seaborn's sns.kdeplot().

Scatter Plots:

Display the relationship between two numerical variables.
Use the plt.scatter() function or Seaborn's sns.scatterplot().

Pair Plots:

Show scatter plots for all pairs of numerical variables in a dataset.
Use Seaborn's sns.pairplot().

Joint Plots:

Combine scatter plots with histograms or KDE plots to visualize the relationship between two numerical variables and their univariate distributions.
Use Seaborn's sns.jointplot().

Heatmaps:

Display the correlation matrix between numerical variables.
Use Seaborn's sns.heatmap().

Line Plots:

Show trends or patterns in numerical data over a continuous variable (e.g., time).
Use the plt.plot() function or Seaborn's sns.lineplot().

ECDF (Empirical Cumulative Distribution Function) Plots:

Visualize the cumulative distribution of a variable.
Use Seaborn's sns.ecdfplot().

Rug Plots:

Display individual data points along an axis, often used in combination with other plots.
Use Seaborn's sns.rugplot().

Swarm Plots:

Combine elements of scatter plots and box plots to show the distribution of data points.
Use Seaborn's sns.swarmplot().

Violin Swarm Plots:

Combine elements of violin plots and swarm plots.
Use Seaborn's sns.violinplot() with the inner='swarm' parameter.

PairGrid:

Create a grid of subplots for pairwise relationships between numerical variables.
Use Seaborn's sns.PairGrid().

Clustermap:

Visualize a matrix dataset with hierarchically-clustered rows and columns.
Use Seaborn's sns.clustermap().

Regplot (Regression Plot):

Display the relationship between two numerical variables and fit a regression line.
Use Seaborn's sns.regplot().

Residual Plots:

Show the residuals (differences between observed and predicted values) in a regression analysis.
Use Seaborn's sns.residplot().

Distribution Plots with Multiple Kernels:

Compare the distribution of a variable using different kernel functions in KDE plots.
Use Seaborn's sns.kdeplot() with the kernel parameter.

Joint Distribution Plot:

Visualize the joint distribution of two numerical variables along with their marginal distributions.
Use Seaborn's sns.jointplot() with the kind='kde' parameter.

Bar Plots (for Aggregated Data):

Display aggregated values (e.g., mean, median) of numerical data using bars.
Use Seaborn's sns.barplot().

###### 2. What are the plots we can create on categorical columns.

Donut Charts:

Stacked Bar Plots:

Grouped Bar Plots:

Compare the values of two categorical variables using grouped bars.

Point Plots:

Box Plots:

Violin Plots:

Swarm Plots:

Heatmaps:

Clustered Bar Plots:

Display the count or aggregated values of two categorical variables using clustered bars.

Mosiac Plots:

Visualize the relationship between two categorical variables using rectangular tiles.

Treemap:

Display hierarchical data as nested rectangles, where each branch represents a categorical variable.

Categorical Box Plots:

Strip Plots:

Swarm Plots (Categorical Variables):

Point Plots (Categorical Variables):
