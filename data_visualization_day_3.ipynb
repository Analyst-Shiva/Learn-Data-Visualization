##### 1. Write a Python program to display a bar chart of the total bill of Gender and write the intuition about the plot.

```python
import matplotlib.pyplot as plt

# Group by gender and calculate total bill
total_bill_gender = data.groupby('sex')['total_bill'].sum().reset_index()

# Plotting the bar chart
plt.bar(total_bill_gender['sex'], total_bill_gender['total_bill'], color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Total Bill')
plt.title('Total Bill Based on Gender')
plt.show()
```

Intuition about the plot:

The bar chart shows the total bill based on gender. The x-axis represents the genders (in this case, 'Male' and 'Female'), and the y-axis represents the total bill amount. Each bar corresponds to the total bill for a particular gender. From the chart, you can compare the total bills between males and females. In this specific example, you can see if there's any significant difference in the total bills spent by males and females.

###### 2. Write a Python program to display a box plot of the total bill of day and hue = "smoker" and write the intuition about the plot.

```python
import seaborn as sns

# Plotting the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=data)
plt.title('Box Plot of Total Bill by Day and Smoker Status')
plt.show()
```

Intuition about the plot:

The box plot allows you to visualize the distribution of total bills for each day, grouped by smoker status. The box represents the interquartile range (IQR), the line inside the box is the median, and the whiskers show the range of the data. Outliers may also be displayed.

By looking at the box plot, you can compare the central tendency and spread of total bills across different days and observe whether there are any noticeable patterns or differences between smokers and non-smokers.

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
# Plotting the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', hue='sex', data=data, split=True)
plt.title('Violin Plot of Total Bill by Day and Gender')
plt.show()
```

Intuition about the plot:

The violin plot displays the distribution of total bills for each day, with separate sections for male and female. Wider areas indicate higher density, and the split allows easy comparison between genders. Peaks show common spending patterns, and outliers highlight unusual spending behavior.

##### 4. Write a Python program to display a strip plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
# Plotting the strip plot
plt.figure(figsize=(10, 6))
sns.stripplot(x='day', y='total_bill', hue='sex', data=data, jitter=True, dodge=True)
plt.title('Strip Plot of Total Bill by Day and Gender')
plt.show()
```

Intuition about the plot:

The strip plot shows individual total bill amounts for each day, with separate points for males and females. Jitter adds randomness to avoid overlap, and dodge separates the plots for easy comparison. It provides a quick view of spending patterns, density, and gender-based differences in total bills.

##### 5. Write a Python program to display heat map of dataframe and write the intuition about the plot.

```python
# Selecting only numeric columns
numeric_columns = data[['total_bill', 'tip']]

# Create a correlation matrix
correlation_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Correlation Matrix (total_bill and tip)')
plt.show()
```

Intuition about the plot:

The heatmap visually represents the correlation between 'total_bill' and 'tip.' Darker colors indicate a stronger relationship. If the color is towards red, there's a positive correlation (both increase together); if towards blue, a negative correlation (one increases, the other decreases). The heatmap aids in quickly identifying patterns and trends between the two variables.

##### 6. Write a Python program to display scatter plot of total bill of tip and write the intuition about the plot.

```python
# Plotting the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['total_bill'], data['tip'])
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

Intuition about the plot:

The scatter plot of 'total_bill' against 'tip' visually displays the relationship between these two variables. Each point represents a combination of a total bill and the corresponding tip. The scatter plot helps quickly identify patterns, trends, and potential outliers in the data, providing a visual understanding of how total bills and tips are related.

##### 7. Write a Python program to display density plot of all numerical columns from the dataframe and write the intution about plot.

```python
# Plotting density plots for all numerical columns
plt.figure(figsize=(10, 6))
for column in data.select_dtypes(include='number').columns:
    sns.kdeplot(data[column], label=column, fill=True)

plt.title('Density Plots of Numerical Columns')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Intuition about the plot:

The density plots illustrate the distribution of values for each numerical column in the dataset. Peaks show common values, while the spread indicates the range of data. Overlapping regions suggest similar distributions, while separation highlights dissimilarities. Outliers may appear as isolated spikes. This visualization provides a quick overview of the variability and patterns in the numerical data.

##### 8. Write a Python program to calculate average total bills by days and display top 4 days by plot.

```python
# Calculate average total bills by days
average_total_bills = data.groupby('day')['total_bill'].mean().sort_values(ascending=False)

# Select top 4 days
top_4_days = average_total_bills.head(4)

# Plotting the bar chart for top 4 days
plt.figure(figsize=(8, 6))
top_4_days.plot(kind='bar', color='skyblue')
plt.title('Top 4 Days by Average Total Bills')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.show()
```

Intuition about the plot:

The bar chart displays the top 4 days with the highest average total bills. The height of each bar represents the average total bill for a specific day, allowing for a quick comparison. This visualization helps identify the most financially significant days in terms of average total spending.

###### 1. What is Histogram chart ?

A histogram is a graphical representation of the distribution of a dataset. It consists of a series of bars, where each bar represents a range of values (or bins), and the height of the bar indicates the frequency or count of data points within that range. Histograms are useful for visualizing the underlying frequency distribution of a continuous variable.

##### 2. What is box plot ?

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum of a dataset. The box represents the interquartile range (IQR) between Q1 and Q3, while the whiskers extend to the minimum and maximum values. Box plots are useful for identifying the central tendency and spread of the data, as well as detecting outliers.

##### 3. What are the plots we can use for univeriate analysis. 

Histograms: Display the distribution of a continuous variable.

Bar Charts: Useful for visualizing the distribution of a categorical variable.

Kernel Density Plots: Estimate the probability density function of a continuous variable.

Box Plots: Illustrate the summary statistics and distribution of a dataset.

##### 4. What are the plots we can use for bivariate analysis.

Scatter Plots: Display the relationship between two continuous variables.
    
Line Charts: Visualize trends over two variables.
    
Heatmaps: Represent the relationship between two categorical variables using color intensity.
    
Box Plots: Compare the distribution of a continuous variable across different categories.

###### 5. What are plot we can use for multivariate analysis.

Pair Plots: Display scatter plots and histograms for multiple variables.

3D Scatter Plots: Visualize relationships in three dimensions.

Bubble Charts: Represent three variables using the size of markers.

Violin Plots: Combine aspects of box plots and kernel density plots for multiple variables.
