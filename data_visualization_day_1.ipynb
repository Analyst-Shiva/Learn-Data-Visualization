##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
import seaborn as sns

# Selecting only numerical columns for the joint plot
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
sns.set(style="whitegrid")
sns.pairplot(data, vars=numerical_columns, hue='Species', markers=["o", "s", "D"], palette="deep")

# Display the plot
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
# Scatter plot with hue="Species"
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=data, palette='Set2')

# Adding labels and title
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Display the plot
plt.show()
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
# Univariate Boxplot for each numerical variable
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df[numerical_columns])
plt.title('Univariate Boxplot for Numerical Variables')

# Bivariate Boxplot with hue="Species"
plt.subplot(1, 2, 2)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Bivariate Boxplot for Sepal Length and Species')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

###### 6. Create a Strip plot 

```python
# Strip plot with hue="Species"
plt.figure(figsize=(8, 6))
sns.stripplot(x='Species', y='SepalLengthCm', data=df, jitter=True, palette='Set2')

# Adding labels and title
plt.title('Strip Plot for Sepal Length with Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

# Display the plot
plt.show()
```

###### 7. Create a Violin Plot.

```python
# Violin plot with hue="Species"
plt.figure(figsize=(10, 6))
sns.violinplot(x='Species', y='SepalLengthCm', data=df, palette='Set2')

# Adding labels and title
plt.title('Violin Plot for Sepal Length with Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

# Display the plot
plt.show()
```

##### 8. Create a pair chart.

##### 9. Create a Heat Map

```python
# Selecting only numerical columns for the correlation matrix
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

# Adding title
plt.title('Correlation Heatmap')

# Display the plot
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
# Selecting only numerical columns for the distribution plot
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create a distribution plot for each numerical column
plt.figure(figsize=(12, 8))
for col in numerical_columns:
    sns.histplot(df[col], kde=True, label=col, bins=20)

# Adding labels and title
plt.title('Distribution Plot for Numerical Columns')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
